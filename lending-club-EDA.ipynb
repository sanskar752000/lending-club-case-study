## Checking the missing or null values

### Checking if any column in the dataset is completely null

```python
# gets and return the sum of all the null column - by default it have axis=0
print('Number of null columns:')
loan_data.isnull().all().sum()

# Result: There are 54 columns that are completely null
```

```python
# lets check what columns are null and make list of them

# this gives the list of all the columns which are completely null
column_with_null_data = list(loan_data.columns[loan_data.isnull().all(axis=0)])
print('List of null value columns:')
column_with_null_data
```

### Removing the null columns from the dataset

```python
# dropping the null columns and saving the result in new dataframe
# here axis=1 represents to delete the columns wise
loan_data_clean = loan_data.drop(column_with_null_data, axis=1)
```

## Checking if any row in the dataset is completely null

```python
# lets get the percentage null value for each column
(loan_data_clean.isna().mean() * 100).round(2)
```

```python
# dropping the columns which have mostly null values(greater than 25%) and will not be useful to us 
list_of_almost_null_columns = ['desc', 'mths_since_last_delinq', 'mths_since_last_record', 'next_pymnt_d']
loan_data_clean.drop(labels=list_of_almost_null_columns, axis=1, inplace=True)
```

```python
# lets check the shape of the dataset again
loan_data_clean.shape

# Result: Now we have only 53 columns left
```

```python
# Now as we have removed completely null and mostly null columns, we can also remove some columns which will
# be not useful to our observation

list_not_required_cols = ['zip_code', 'url', 'initial_list_status', 'application_type', 'acc_now_delinq', 'title', 
                          'chargeoff_within_12_mths', 'delinq_amnt', 'tax_liens', 'pymnt_plan', 'policy_code', 
                          'collections_12_mths_ex_med', 'revol_bal', 'revol_util', 'pub_rec', 'delinq_2yrs', 'total_acc',
                         'out_prncp', 'total_pymnt', 'out_prncp_inv', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int',
                         'total_rec_late_fee', 'recoveries', 'collection_recovery_fee', 'last_pymnt_d', 'last_pymnt_amnt',
                         'last_credit_pull_d', 'inq_last_6mths', 'open_acc', 'earliest_cr_line', 'pub_rec_bankruptcies']

# dropping the above list of columns from the dataset
loan_data_clean.drop(labels=list_not_required_cols, axis=1, inplace=True)
```

```python
# emp_title is one of the field which we see is not very useful for our analysis as there are multiple different titles 
#which won't make any useful pattern for our aim.

# So dropping the emp_title from the dataset
loan_data_clean.drop('emp_title', axis=1, inplace=True)

# Also we can remove id and member_id columns as they are unqiue identifiers and are not useful for our analysis.
loan_data_clean.drop(labels=['id', 'member_id'], axis=1, inplace=True)

# Checking the shape again
loan_data_clean.shape
```

## Imputing and formatting the columns

```python
# Now let's check the info all the columns we're left with
loan_data_clean.info()

# Result: We have all the columns with there datatypes
```

```python
# Now checking with box plot that what's the range of annual income we have for each category of emp_length this will not show NaN category
sns.boxplot(x=loan_data_clean.emp_length, y=loan_data_clean.annual_inc, data=loan_data_clean)

plt.xticks(rotation=90)
plt.show()

# Result: As we can see from the boxplot that there are huge outliers present for this data
```

```python
# Let's look into the annual income of these NaN records in the emp_length
loan_data_clean[loan_data_clean.emp_length.isna()].annual_inc.sort_values(ascending=False).unique()

# Result: we can see that we cannot correctly predict the what's the correct experience for these values but the annual 
# income for the NaN experience record is not significantly high compared to the other categories of emp_length so, we can make
# an assumption here to give 0 value to all the NaN values in emp_length
```

```python
# We can format emp_length correctly as it is a useful column for our analysis
print('emp_length before: ')
print(loan_data_clean.emp_length.head())

# we can also see from above observation that some percentage of emp_length are have null values so we will fill them with 0
# as per the above assumption made.
loan_data_clean.emp_length.fillna('0', inplace=True)

# Now converting the type and fixing the formatting for the emp_length

''' 
Here's an another assumption that there are multiple person who have <1 year and 1 year of distinguish experience
which we are considering as 1 year whole because people who have maybe 8-10 months of experience won't make much difference 
in terms of having 12 months(1 year) of experience.
And there were multiple records which were null (aroung 2.7% which is not very huge) are given 0 years of experience.
'''
loan_data_clean.emp_length = loan_data_clean.emp_length.str.extract('(\d+)')
print('emp_length after: ')
print(loan_data_clean.emp_length.head())
```

```python
# We can see int_rate is of type object as it have %, so let's convert this to float type by removing the %
print('int_rate before: ')
print(loan_data_clean.int_rate.head())

loan_data_clean.int_rate = loan_data_clean.int_rate.apply(lambda x: float(x.rstrip('%')))
print('int_rate after: ')
print(loan_data_clean.int_rate.head())
```

### Deriving new columns

```python
# Let's convert the issue_date from object to datetime datatype
print('issue_d info before: ')
loan_data_clean.issue_d.info()
loan_data_clean.issue_d = pd.to_datetime(loan_data_clean.issue_d, format='%b-%y')

print()

print('issue_d info after: ')
print(loan_data_clean.issue_d.info())

print()

print(loan_data_clean.issue_d.head())
# Result: Converted the issue_date in yyyy-mm-dd format of type datetime
```

## Outliers detection and treatment

```python
# L`et's start by checking the outliers for annual income column first.
# Note: Above while treating the null values we have seen in the boxplot, for annual income respective to different category
# of employee experience length, that there are significantly high outliers are present for the same. 
sns.boxplot(data=loan_data_clean.annual_inc)
plt.title('Annual Income Boxplot', fontsize=14, color='b')
plt.ylabel('Annual Income')
plt.show()

# Result: We can see there are very large outliers present for this Annual income
```

```python
# Lets describe annual income column to get insight of it
print('Annual Income description: ')
print(loan_data_clean.annual_inc.describe())

print()

# And Lets see the percentile more closely for annual income attribute
print('Percentiles for Annual Income: ')
print(loan_data_clean.annual_inc.quantile([0.75, 0.80, 0.85, 0.90, 0.95, 0.99, 1]))

# Result: We can see that there is huge difference jump of value from 0.99 to 1 percentile, because of which we're having huge raise in the
# mean of the dataset, so in the next step let's remove the rows above 0.99 percentile
```

```python
#Removing the rows above the 0.99 percentile for annual income column
threshold_percentile = loan_data_clean.annual_inc.quantile(0.99)
loan_data_clean = loan_data_clean[loan_data_clean.annual_inc <= threshold_percentile]

# As we have deleted some rows from the dataset we need to reset the index for the same
loan_data_clean = loan_data_clean.reset_index(drop=True)
```

```python
# Let's analyse the purpose column now
print((loan_data_charged_off.purpose.value_counts()*100)/len(loan_data_clean))


sns.countplot(x=loan_data_charged_off.purpose, data=loan_data_clean)
plt.xticks(rotation=90)
plt.show()

# Result: Not very superisengly, borrower have taken loan for the purpose of debt_consolidation are very
# risky and most likely to default. The major reason is that these people take loan for paying other debts
# they have and then over spend the money which cause them huge finance failure and lead them to being default.
# Another interesting reason can be, the borrowers who take loans for debt consolidation purpose often have
# less credit score and because of which they get this type of loan at very high interese rates, which make 
# it diffcult for them to pay.
```
