```python

# Based on the above observation and according to our aim we need to see what's the deriving features 
# between the loans that have been fully paid or charged off. So will not consider loans that have status current.
loan_data_final = loan_data_clean[loan_data_clean.loan_status.isin(['Charged Off', 'Fully Paid'])]

loan_data_final = loan_data_final.reset_index(drop=True)

# let's check the shape for the dataset for charged off, to see how much record we're left with
loan_data_final.shape
```

```python
# Let's build a custom method to print two plot separately side by side
# The left one will be the percentage of a categorical variable in the entire dataset and the right one
# will be the percentage charged off loans

def build_percentage_plot(dataframe, by, sortByIndex=False):
    plt.subplot(1, 2, 1)
    values = (loan_data_final[by].value_counts(normalize=True)*100).round(2)
    if sortByIndex:
        values = values.sort_index()
    ax = values.plot.bar(color=sns.color_palette('husl', 16))
    ax.bar_label(ax.containers[0], label_type='edge')
    ax.set_ylabel("% in dataset", fontsize=16)
    ax.set_xlabel(by, fontsize=14)

    plt.subplot(1, 2, 2)
    values = (loan_data_final.loc[loan_data_final.loan_status == 'Charged Off'][by].value_counts(normalize=True)*100).round(2)
    if sortByIndex:
        values = values.sort_index()
    ax = values.plot.bar(color=sns.color_palette('husl', 16))
    ax.bar_label(ax.containers[0], label_type='edge')
    ax.set_ylabel("% in charged off loans", fontsize=16)
```

```python
# Let's checkout verification_status and what insights it have

fig, ax = plt.subplots(figsize=(18, 9))
build_percentage_plot(loan_data_final, 'verification_status')
plt.title('% Verification Status Barplot(Charged Off)', fontsize=14, color='b')
plt.show()

# Observation: It is pretty clear from the output that borrower who are Not Verified are most likely to
# default, that we can see in the plot at right side with about 38.14% of defaulters, can be mark risky as they could not be a genuine person. 
# But among charged off loans, it is interesting to see that percentage of verified category rises to 36.39%, 
# this behaviour is unexpected and will investigated in bivariate analysis.
```

```python
# Function to format and show the labels. Took some reference from https://stackoverflow.com/questions/28931224/how-to-add-value-labels-on-a-bar-chart/48372659#48372659 
# to format the labelling of plots.
def formatAndShowLabels(ax, dataD):
    rects = ax.patches
    locs, labels = plt.xticks()
    i=0
    counts = {}
    for key, value in dataD.items():
        counts[str(key)] = value

    for rect in rects:
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = 6
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'
        
        try:
            label = "{:.1f}%".format(y_value) + "\ncnt.-> " + str(counts[str(labels[i].get_text())])
        except:
            label = "{:.1f}%".format(y_value)
        
        i = i+1

        # Create annotation
        plt.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(0, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for
                                        # positive and negative values.
```

```python

# Let's now define a function here which basically creates a bar plot for Charged Off category
# The plots will be in percentage - we are basically going to form a percentage for each category out of the 
# total charged off.
# If it's a continuous column we will form the bucket/bin so to form the category for that column.

def plotBarOnChargedOffLoanStatus(dataframe, by, loanstatus='Charged Off'):
    grp = dataframe.groupby(['loan_status',by])[by].count()
    cnt = dataframe.groupby(by)[by].count()

    percentages = (grp.unstack() * 100 / cnt.T)

    ax = percentages.loc[loanstatus].plot.bar(color=sns.color_palette('husl', 16))
    ax.set_ylabel('% of loans ' + loanstatus)
    plt.margins(0.2, 0.2)
    formatAndShowLabels(ax, grp[loanstatus].to_dict())
    plt.tight_layout()
    return ax
```

```python
# Let's start with our segmented analysis and for that let's pickup purpose column as it's the important deriving factor for defaulters

plt.figure(figsize=(15, 7))
plotBarOnChargedOffLoanStatus(loan_data_final, 'purpose')
plt.title('% Purpose(Charged Off)', fontsize=14, color='b')
plt.show()

# Observation: From the the outcome we can derive that,
# - 27.4 % of the loans taken for the purpose for small_business is charged off, which makes it most risky purpose. This is also noticed above while doing categorical univariate analysis as there was spike
#   in percentage of small_business category
```

```python
# Now let's check how much emp_length of borrower caused more defaults.

plt.figure(figsize=(15, 7))
plotBarOnChargedOffLoanStatus(loan_data_final, 'emp_length')
plt.title('% Employee Exp.(Charged Off)', fontsize=14, color='b')
plt.show()

# Observation: We can clearly and observed in above analysis also that borrower having 0 years of experience causes 
# more defaults followed by 10 years of experience. So Lending Club should prevent giving loans to the people 
# who have less or 0 year experience as they are most likely to be defaulted.
```

7. They should reduce the loans taken with higher interest for 60 months, because these are likely to get default.
